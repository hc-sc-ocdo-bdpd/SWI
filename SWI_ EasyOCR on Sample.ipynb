```python
# commencing with easyocr
reader = easyocr.Reader(['en', 'fr'])
```

```python
from docx import Document

def read_docx(file_name: str):
    '''
    extracts the text from a docx file in the "dbfs:/FileStore/noc_pm_pdfs/" folder
    '''
    file_path = "/dbfs/FileStore/noc_pm_pdfs/" + file_name
    doc = Document(file_path)
    text = []
    for para in doc.paragraphs:
        text.append(para.text)
    return "\n".join(text)
```

```python
from PIL import Image, ImageDraw, ImageFont

def text_to_image(text, output_path: str):
    # Create a blank image with white background
    image = Image.new("RGB", (800, 600), "white")
    draw = ImageDraw.Draw(image)
    
    # Use a default font (adjust path if needed)
    try:
        font = ImageFont.truetype("arial.ttf", 14)
    except IOError:
        font = ImageFont.load_default()
    
    # Draw text on the image
    draw.text((10, 10), text, fill="black", font=font)
    
    # Save the image
    image.save(output_path)
    print(f"Image saved as {output_path}")
```

```python
def docx_to_png(docx_name:str, output_path: str):
    text = read_docx(docx_name)
    text_to_image(text, output_path)
```

```python
# function to run easyocr
import os

def easyocr_doc(path: str) -> list:
    '''
    takes path to pdf/docx file and runs EasyOCR on the document, effectively returning a list of strings
    TODO: flatten output to one long string
    '''
    print(f"[EASYOCR] Extracting Text")
    root, ext = os.path.splitext(path) # sending path extension to lowercase
    path = root + ext.lower()
    try:
        if path.endswith(".docx"):
           # TODO: implement function to convert docx to png
           pass
        elif path.endswith('.png') or path.endswith('.jpg'):
            pass
        else:
            raise Exception("File type not supported")
        result = reader.readtext(path, detail=0)
        return result
        
    except Exception as e:
        print(f"[OCR ERROR] {path} -> {e}")
        return ""
```

```python
# data structure for output of ocr

from pyspark.sql.types import StructType, StructField, StringType

schema = StructType([
    StructField("path", StringType(), True),
    StructField("final_text", StringType(), True)
])
```

```python
pdfs = dbutils.fs.ls("/FileStore/accuracy_experiment/")

results = []
for pdf in pdfs:
    path = "/dbfs" + pdf.path[len("dbfs:"):]
    results.append({'path': path, 'final_text': (easyocr_doc(path))})
```

```python
# creating Spark dataframe
pdf_df = spark.createDataFrame(results, schema)
pdf_df.show()
pdf_df.write.mode("overwrite").saveAsTable("easyocr_exp")
```
